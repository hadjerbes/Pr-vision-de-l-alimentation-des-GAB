## Analyse de normalité

### Caractéristiques à calculer pour vérifier la normalité:

On va calculer le Skewness et le Kurtosis :

-Skewness (asymétrie): mesure l'asymétrie d'une distribution d'une variable aléatoire.

-Kurtosis (aplatissement): mesure la queue d'une distribution d'une variable aléatoire.

Dans ce qui suivra je vais utilisé le kurtosis normalisé de fisher.

```python
from scipy.stats import skew, kurtosis

# Calculer la skewness (asymétrie) des variables numeriques
skewness_retrait_sem = skew(df['retrait_Semaine'])
skewness_Nbr_Transaction_Semaine = skew(df['Nbr_Transaction_Semaine'])
skewness_retrait_Weekend = skew(df['retrait_Weekend'])
skewness_Nbr_Transaction_Weekend = skew(df['Nbr_Transaction_Weekend'])
skewness_retrait_evenement = skew(df['retrait_evenement'])
skewness_nbr_transaction_evenement = skew(df['nbr_transaction_evenement'])

# Calculer le kurtosis normalisé (aplatissement) des variables numeriques
kurtosis_retrait_sem = kurtosis(df['retrait_Semaine'],fisher=True)
kurtosis_Nbr_Transaction_Semaine = kurtosis(df['Nbr_Transaction_Semaine'],fisher=True)
kurtosis_retrait_Weekend = kurtosis(df['retrait_Weekend'],fisher=True)
kurtosis_Nbr_Transaction_Weekend = kurtosis(df['Nbr_Transaction_Weekend'],fisher=True)
kurtosis_retrait_evenement = kurtosis(df['retrait_evenement'],fisher=True)
kurtosis_nbr_transaction_evenement = kurtosis(df['nbr_transaction_evenement'],fisher=True)

print("Skewness (asymétrie) de retrait_Semaine:", skewness_retrait_sem)
print("Kurtosis (aplatissement) de retrait_Semaine:", kurtosis_retrait_sem)
print("Skewness (asymétrie) de Nbr_Transaction_Semaine:", skewness_Nbr_Transaction_Semaine)
print("Kurtosis (aplatissement) de Nbr_Transaction_Semaine:", kurtosis_Nbr_Transaction_Semaine)
print("Skewness (asymétrie) de retrait_Weekend:", skewness_retrait_Weekend)
print("Kurtosis (aplatissement) de retrait_Weekend:", kurtosis_retrait_Weekend)
print("Skewness (asymétrie) de Nbr_Transaction_Weekend:", skewness_Nbr_Transaction_Weekend)
print("Kurtosis (aplatissement) de Nbr_Transaction_Weekend:", kurtosis_Nbr_Transaction_Weekend)
print("Skewness (asymétrie) de retrait_evenement:", skewness_retrait_evenement)
print("Kurtosis (aplatissement) de retrait_evenement:", kurtosis_retrait_evenement)
print("Skewness (asymétrie) de nbr_transaction_evenement:", skewness_nbr_transaction_evenement)
print("Kurtosis (aplatissement) de nbr_transaction_evenement:", kurtosis_nbr_transaction_evenement)
```

pour la variable 'retrait_Semaine':

-skewness=2.11 elle est positive cela indique une asymétrie vers la droite. la majorité des valeurs sont concentrées à gauche et qu'il y a quelques valeurs elevées à droite.

-kurtosis=6.69 > 0 , cela signifie que sa distribution est rétrécie, les valeurs sont plus regroupées vers la moyenne.

##### on peut dire que la distribution de la variable 'retrait semaine' n'est pas normale.

### Les tests statistiques:

J'ai utilisé le test de kolmogorov car il convient aux echantillons de taille importantes.

```python
retrait_evenement = df["retrait_Semaine"]

# Effectuez le test de Kolmogorov-Smirnov
ks_statistic, p_value = stats.kstest(retrait_evenement, 'norm')

# Affichez les résultats
print(f"Statistique de test de Kolmogorov-Smirnov : {ks_statistic:.4f}")
print(f"P-valeur : {p_value:.4f}")

# Interprétation
alpha = 0.05
if p_value < alpha:
    print("La distribution n'est pas normale (rejeter l'hypothèse nulle).")
else:
    print("La distribution suit une loi normale (ne pas rejeter l'hypothèse nulle).")
```
